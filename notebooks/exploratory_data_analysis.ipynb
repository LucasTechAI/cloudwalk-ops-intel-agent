## Functions 

```python
def show_bars_plot(
    data: pd.DataFrame,
    x_col: str,
    y_col: str,
    title: str,
    xlabel: str = None,
    ylabel: str = None,
    figsize: tuple = (10, 6)
) -> None:
    """
    Displays a horizontal bar chart sorted by y_col, with a light gray vertical grid.
    """

    # --- Validations
    if x_col not in data.columns:
        raise ValueError(f"Column '{x_col}' not found in DataFrame.")
    if y_col not in data.columns:
        raise ValueError(f"Column '{y_col}' not found in DataFrame.")
    if data.empty:
        raise ValueError("Input DataFrame is empty.")

    # --- Sorting
    df_sorted = data.sort_values(by=y_col, ascending=True)

    # --- Color gradient (light → dark)
    n = len(df_sorted)
    colors = plt.cm.Blues(np.linspace(0.4, 1.0, n))

    # --- Plot
    fig, ax = plt.subplots(figsize=figsize)
    ax.barh(df_sorted[x_col], df_sorted[y_col], color=colors)

    # --- Grid vertical (x-axis)
    ax.xaxis.grid(True, linestyle='--', linewidth=0.6, color='lightgray')
    ax.yaxis.grid(False)

    # --- Titles & labels
    ax.set_title(title)
    if xlabel:
        ax.set_xlabel(xlabel)
    if ylabel:
        ax.set_ylabel(ylabel)

    plt.tight_layout()
    plt.show()

    
def show_histogram(
    data: pd.DataFrame,
    col: str,
    title: str = None,
    xlabel: str = None,
    ylabel: str = "Frequency",
    bins: int = 30,
    figsize: tuple = (10, 6),
    color: str = 'steelblue'
) -> None:
    """
    Displays a histogram for numeric (int/float) columns with distribution statistics.
    
    Parameters:
    -----------
    data : pd.DataFrame
        Input DataFrame
    col : str
        Column name to plot (must be numeric)
    title : str, optional
        Plot title (defaults to "Distribution of {col}")
    xlabel : str, optional
        X-axis label (defaults to column name)
    ylabel : str, optional
        Y-axis label (default: "Frequency")
    bins : int, optional
        Number of bins (default: 30)
    figsize : tuple, optional
        Figure size (default: (10, 6))
    color : str, optional
        Bar color (default: 'steelblue')
    """
    
    # --- Validations
    if col not in data.columns:
        raise ValueError(f"Column '{col}' not found in DataFrame.")
    if data.empty:
        raise ValueError("Input DataFrame is empty.")
    if not pd.api.types.is_numeric_dtype(data[col]):
        raise ValueError(f"Column '{col}' must be numeric (int or float).")
    
    # Remove NaN values
    values = data[col].dropna()
    
    if len(values) == 0:
        raise ValueError(f"Column '{col}' has no valid numeric values.")
    
    # --- Calculate statistics
    mean_val = values.mean()
    median_val = values.median()
    std_val = values.std()
    max_val = values.max()
    min_val = values.min()
    
    # --- Plot
    fig, ax = plt.subplots(figsize=figsize)
    
    # Histogram
    n, bins_edges, patches = ax.hist(values, bins=bins, color=color, alpha=0.7, edgecolor='black')
    
    # Add vertical lines for mean and median
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
    
    # --- Grid
    ax.yaxis.grid(True, linestyle='--', linewidth=0.6, color='lightgray', alpha=0.7)
    ax.xaxis.grid(False)
    
    # --- Titles & labels
    ax.set_title(title or f"Distribution of {col}")
    ax.set_xlabel(xlabel or col)
    ax.set_ylabel(ylabel)
    ax.legend()
    
    # Add statistics text box
    #add max min
    stats_text = f'Count: {len(values)}\nMax: {max_val:.2f}\nMin: {min_val:.2f}\nStd: {std_val:.2f}'
    ax.text(0.98, 0.98, stats_text, transform=ax.transAxes, 
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.show()
```
